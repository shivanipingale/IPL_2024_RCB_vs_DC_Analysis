**Import the necessary libraries I will use for this project**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Reading The data**

```python
df = pd.read_csv(r"IPL 2024 RCB vs DC.csv")
df
```

**Check the Data**

```python
df.head()
```

**Get Summary of Data**

```python
df.info()
```

**Delete Unnecessary Columns**

```python
df.drop(columns = 'fielders', inplace=True)
df
```

**Display All Rows**

```python
pd.options.display.max_rows = 247
df
```

**Get details of numeric values**

```python
df.describe()
```

**Check Null Values**

```python
df.isnull().sum()
```

**Check for the players who runs most**

```python
runs_by_player = df.groupby(['team','batter'])['runs_batter'].sum().sort_values(ascending = False)
runs_by_player
```

**Check top 10 Scorers of the match with bar chart**

```python
runs_by_player.head(10).sort_values(ascending=True).plot(kind = 'barh', color = ['darkblue', 'teal', 'salmon', 'orange'])
plt.xlabel('Total Runs')
plt.ylabel('Player')
plt.title('Top 10 Run scorers')
```

**AR Patel (57 runs) from DC** and **RM Patidar (52 runs) from RCB** crossed the half-century mark !!
Contributing significantly to the team’s total.

**Player who hits more 6s and 4s**

```python
more_six = df[df['runs_batter'] == 6].groupby('batter')['runs_batter'].count()
more_six
```

```python
more_four = df[df['runs_batter'] == 4].groupby('batter')['runs_batter'].count()
more_four
```

```python
Most_boundaries = pd.merge(more_six, more_four, how = 'outer', on = 'batter')
Most_boundaries.fillna(00 , inplace = True)
Most_boundaries.rename(columns = {'runs_batter_x' : 'Sixes', 'runs_batter_y' : 'Fours'}).plot(kind = 'barh', stacked = True, color = ['darkblue', 'red'])
plt.xlabel('Frequency of 6s and 4s')
plt.ylabel('Batter')
plt.title('Most Boundaries')
```

From above graph, we see **most boundaries hits by AR Patel** with **5-4s and 3-6s**. 

**Check the correlation between numeric data**

```python
corr = df.corr(numeric_only=True)
corr
```

1. -0.17 correlation for over and runs_batter - As the over number increases, the runs scored per ball by the batter slightly decrease.
2. 0.989 for runs_batter and runs_total - Its positive, because there is strong impact of batters in the total runs

**Show the result using Hexbin chart**

```python
plt.hexbin(df['over'], df['runs_batter'], gridsize=20, cmap = 'YlGn')
plt.xticks(df['over'])
plt.yticks(df['runs_batter'])
plt.ylabel("Runs by Batter")
plt.xlabel("Overs")
plt.title("Runs Scored by Batter per Over")
cbar = plt.colorbar() 
cbar.set_label("Frequency of Runs") 
```

From above hexbin graph, cleary visualize that at early overs, most of the scoring is concentrated at lower run values (0-2 runs per batter),few high-run hexagons indicate occasional boundaries (4s and 6s).


At middle overs(6 - 15), consistent spread of runs, especially around 1-3 runs per batter.

**Frequency of Runs by Batter**

```python
df['runs_batter'].value_counts().plot(kind = 'bar', color = 'darkblue')
plt.xlabel('Runs by Batter')
plt.ylabel('Frequency of Runs')
plt.title("Frequency of Runs by Batter")
```

**Top 10 Batter performance by over with more runs**

```python
most_run_by_over = df.groupby(['team','over', 'batter'])['runs_total'].sum().sort_values(ascending=True).tail(10)
plt.figure(figsize=(9,6))
most_run_by_over.plot(kind = 'barh', xlabel = 'Total Runs', ylabel = 'Over per Batter', color = ['darkblue', 'teal', 'salmon', 'orange'])
plt.title("Top 10 Runs by Over")
```

From above graph we can see that **C Green** from RCB scored highest runs in 16th over that play a vital role for the team. 


**Kind of Wicket by Bowler**

```python
wickets_data = df['wicket_kind'].value_counts()
wickets_data.plot(kind = 'bar',color = ['darkblue', 'teal', 'salmon', 'orange'])
plt.xlabel('Type of Wicket')
plt.ylabel('Player_out')
plt.title('Kind of Wicket')
```

From above chart we see that, most of the wickets done by caught.

**Check the Bowler performance by over with total runs given and wicket kind**

```python
Bowler_performance = df.groupby(['over','bowler']).agg({'runs_total': 'sum','runs_extras':'sum','wicket_kind': 'count'}).sort_values(by = 'over',ascending=True)
Bowler_performance
```

```python
Wickets_by_bowler = df.groupby('bowler')['wicket_kind'].count().sort_values(ascending =False)
Wickets_by_bowler.plot(kind = 'bar', color = ['red', 'darkblue'])
plt.title('Top player in Bowling')
plt.xlabel('Bowler')
plt.ylabel('Wickets')
```

From above bowler data, **Top performer 'Yash Dayal'** (From RCB) **conceded only 20 runs with 4 wickets**, while **Kuldeep Yadav** (Bowler of DC) **conceded 53 (includes 2 extra runs)** runs to RCB, which gave a significant advantage to RCB in building a strong total. 

**Performance of the team by each over**

```python
performance_by_RCB = df[df['team'] == 'Royal Challengers Bengaluru'].groupby('over')['runs_total'].sum()
performance_by_RCB
```

```python
performance_by_DC = df[df['team'] == 'Delhi Capitals'].groupby('over')['runs_total'].sum()
performance_by_DC
```

```python
RCB_DC_performance = pd.merge(performance_by_DC, performance_by_RCB, how = 'outer', on = 'over')
RCB_DC_performance = RCB_DC_performance.rename(columns = {'runs_total_x':'DC','runs_total_y':'RCB'})
RCB_DC_performance
```

```python
plt.figure(figsize=(20,10))
RCB_DC_performance.plot(kind = 'line', xlabel = 'Over', ylabel = 'Runs by team', title = 'Performance of team per over', xticks = df['over'], color = ['blue', 'red'])
```

**Most players out with extra runs got**

```python
team = df.groupby('team').agg({'runs_extras': 'sum','player_out': 'count'})
team = team.rename(columns = {'runs_extras': 'Extra Runs', 'player_out': 'Players Out'})
team
```
