```python
runs_by_player = df.groupby(['team','batter'])['runs_batter'].sum().sort_values(ascending = False)
runs_by_player
```

```python
runs_by_player.head(10).sort_values(ascending=True).plot(kind = 'barh', color = ['darkblue', 'red'])
plt.xlabel('Total Runs')
plt.ylabel('Team-Player')
plt.title('Runs scored by player')
```

```python
more_six = df[df['runs_batter'] == 6].groupby('batter')['runs_batter'].count()
more_six
```

```python
more_four = df[df['runs_batter'] == 4].groupby('batter')['runs_batter'].count()
more_four
```

```python
Most_boundaries = pd.merge(more_six, more_four, how = 'outer', on = 'batter')
Most_boundaries.fillna(00 , inplace = True)
Most_boundaries.rename(columns = {'runs_batter_x' : 'Sixes', 'runs_batter_y' : 'Fours'}).plot(kind = 'barh', stacked = True, color = ['red', 'darkblue'])
```

**AR Patel (57 runs) from DC** and **RM Patidar (52 runs) from RCB** crossed the half-century mark !!
Contributing significantly to the teams total.

```python
plt.hexbin(df['over'], df['runs_batter'], gridsize=20, cmap = 'YlGn')
plt.xticks(df['over'])
plt.yticks(df['runs_batter'])
plt.ylabel("Runs by Batter")
plt.xlabel("Overs")
plt.title("Runs Scored by Batter per Over")
cbar = plt.colorbar() 
cbar.set_label("Frequency of Runs") 
```

From above hexbin graph, cleary visualize that at early overs, most of the scoring is concentrated at lower run values (0-2 runs per batter),few high-run hexagons indicate occasional boundaries (4s and 6s).


At middle overs(6 - 15), consistent spread of runs, especially around 1-3 runs per batter.

```python
df['runs_batter'].value_counts().plot(kind = 'bar', color = 'salmon')
plt.xlabel('Runs by Batter')
plt.ylabel('Frequency of Runs')
plt.title("Frequency of Runs by Batter")
```

```python
most_run_by_over = df.groupby(['team','over', 'batter'])['runs_total'].sum().sort_values(ascending=False).head(15)
plt.figure(figsize=(9,6))
most_run_by_over.plot(kind = 'barh', xlabel = 'Total Runs', ylabel = 'Over per batter' , title = 'Most runs by Over', 
                      color = [ 'lightgreen', 'darkblue', 'cyan'])
```

From above graph we can see that **C Green** from RCB scored highest runs in 16th over.  


```python
wickets_data = df['wicket_kind'].value_counts()
wickets_data.plot(kind = 'bar',color = ['yellow', 'orange', 'green', 'blue'])
plt.xlabel('Type of Wicket')
plt.ylabel('Player_out')
plt.title('Kind of Wicket')
```

```python
Bowler_performance = df.groupby(['over','bowler']).agg({'runs_total': 'sum','runs_extras':'sum','wicket_kind': 'count'}).sort_values(by = 'over',ascending=True)
Bowler_performance
```

```python
Wickets_by_bowler = df.groupby('bowler')['wicket_kind'].count().sort_values(ascending =False)
Wickets_by_bowler.plot(kind = 'bar', color = 'lightgreen')
```

From above bowler data, **Top performer 'Yash Dayal'** (From RCB) **conceded only 20 runs with 4 wickets**, while **Kuldeep Yadav** (Bowler of DC) **conceded 53 (includes 2 extra runs)** runs to RCB, which gave a significant advantage to RCB in building a strong total. 

```python
performance_by_RCB = df[df['team'] == 'Royal Challengers Bengaluru'].groupby('over')['runs_total'].sum()
performance_by_RCB
```

```python
performance_by_DC = df[df['team'] == 'Delhi Capitals'].groupby('over')['runs_total'].sum()
performance_by_DC
```

```python
RCB_DC_performance = pd.merge(performance_by_DC, performance_by_RCB, how = 'outer', on = 'over')
RCB_DC_performance = RCB_DC_performance.rename(columns = {'runs_total_x':'DC','runs_total_y':'RCB'})
RCB_DC_performance
```

```python
team = df.groupby('team').agg({'runs_extras': 'sum','player_out': 'count'})
team = team.rename(columns = {'runs_extras': 'Extra Runs', 'player_out': 'Players Out'})
team.plot(kind = 'bar', color = ['red', 'darkblue'])
plt.xticks(rotation=0)
```

From above graph we see that RCB got advantage from 10 extra runs also. 

```python
plt.figure(figsize=(20,10))
RCB_DC_performance.plot(kind = 'line', xlabel = 'Over', ylabel = 'Runs by team', title = 'Performance of team per over', xticks = df['over'], color = ['blue', 'red'])
```

```python
total_run_by_team = df.groupby(['team'])['runs_total'].sum()
total_run_by_team
```

RCB scored the highest runs in the 16th over, showing an aggressive phase while DC had a more fluctuating run rate, with a peak around the 12th over but from the 15th over onward, DCs scoring rate dropped sharply, indicating difficulty in finishing strong.
